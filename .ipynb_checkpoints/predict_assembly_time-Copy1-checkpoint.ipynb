# Прогнозирование времени сборки онлайн-заказа в магазине разных форматов

Параметры датасета:
- `assembly_id` - идентификатор сборки заказа;
- `unique_goods_amount` - кол-во уникальных товаров в заказе;
- `pieces_goods_amount` - кол-во уникальных штучных товаров;
- `kg_goods_amount` - кол-во уникальных весовых товаров;
- `pieces_goods_total` - общее кол-во штучных товаров;
- `kg_goods_weight` - общий вес весовых товаров;
- `store_format` - формат торгового объекта;
- `start_assembly` - дата и время начала сборки заказа; 
- `end_assembly` - дата и время окончания сборки заказа;
- **`assembly_time`** - продолжительность сборки заказа (целевая переменная, таргет).

Задача: 

**Построить модель, разбив выборку на тренировочную и тестовую.**

## 1. Обработка данных

### Чтение данных

```python
# смотрим как прочитались данные
df = pd.read_csv('data-0104-3004.csv')
df.head()
```

```python
# смотрим размерность датафрейма: (строки, колонки)
df.shape
```

```python
# смотрим есть ли пропуски данных, какие типы данных присвоены
df.info()
```

```python
weight_goods = {
    'kg_goods_amount': 'weight_goods_amount',
    'kg_goods_weight': 'weight_goods_total_weight',
    'pieces_goods_total': 'pieces_goods_total_count'
}
df.rename(weight_goods, axis=1, inplace=True)
df.head()
```

```python
df['pieces_goods_total_count'].value_counts()
```

### Вычисление таргета

```python
# смотрим какие типы данных присвоены
df.info()
```

```python
# приведем тип данных для столбцов со временем начала и окончания сборки 
df['start_assembly'] = pd.to_datetime(df['start_assembly'])
df['end_assembly'] = pd.to_datetime(df['end_assembly'])

df.info()
```

```python
# рассчитаем время сборки заказа
df['assembly_time'] = df['end_assembly'] - df['start_assembly']
df.head()
```

```python
# приведем время сборки в адсолютному значению в минутах
convert_to_sec = lambda x: x.total_seconds() / 60
df['assembly_time'] = df['assembly_time'].apply(convert_to_sec)
df
```

```python
# избавимся от лишних столбцов, поменяем порядок
df = df.drop(['assembly_id', 'start_assembly', 'end_assembly'], axis=1)
df = df[list(reversed(df.columns.tolist()))]
df
```

### Отбор форматов магазинов

```python
# посмотрим как распределяется время сборки по форматам ТО
func_agg = {
    'assembly_time': ['mean', 'std', 'min', 'max', 'count']
}
df.groupby('store_format').agg(func_agg).sort_values(by=[('assembly_time', 'count')], ascending=False)
```

```python
df['store_format'].nunique()
```

```python
# зададим минимальное число заказов для каждого формата магазина
df['store_format'].value_counts()>1000
```

```python
df['store_format'].unique().tolist()[:14]
```

```python
df=df.loc[df['store_format'].isin(df['store_format'].unique().tolist()[:14])]
df
```

### Поиск пропущенных значений

```python
# проверим, что столбцы не содержат никаких пустых значений 
df.isnull().sum()
```

```python
# удалим строки содержащие пустые значения
df.dropna(axis=0, inplace=True)
```

```python
# проверим, какие столбцы содержат пустые значения и сколько их 
df.isnull().sum()
```

```python
def plot_hist_box(data):
    plt.figure(figsize=(15, 8))
    plt.subplot(2, 1, 1)
    sns.histplot(data=data, bins=45, kde=True)
    plt.grid()

    plt.subplot(2, 1, 2)
    sns.boxplot(data=data, orient='horizontal')
    plt.grid()

    plt.show()
```

## 3. Обучение моделей

### Разделение данных на тренировочную и тестовую выборки

```python
from sklearn.model_selection import train_test_split
```

### Линейные модели

```python
data1 = df2.copy()
data1
```

```python
# data = data[data.assembly_time<30]
data1 = data1[data1['store_format_Гипермаркет']==1]
data1 = data1[data1.columns.tolist()[:3]]
data1
```

```python
# Разделим датафрейм на тестовую и тренировочную выборки
X = data1.drop('assembly_time', axis=1)
y = data1['assembly_time']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# выполним нормализацию всех признаков
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Линейная регрессия

```python
# импортируем модель и метрики оценки качества обучения
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# обучаем модель
model_LR = LinearRegression()
model_LR.fit(X_train, y_train)
```

```python
# предсказываем таргет
y_pred_LR = model_LR.predict(X_test)
y_pred_LR
```

```python
# оцениваем качество обученной модели
mean_absolute_error(y_pred_LR, y_test)
mean_squared_error(y_pred_LR, y_test)

# print('Train MAE: ', mean_absolute_error(y_train, y_train_predicted))
print('Test MAE: ', mean_absolute_error(y_test, y_pred_LR))

# print('Train MSE: ', mean_squared_error(y_train, y_train_predicted))
print('Test MSE: ', mean_squared_error(y_test, y_pred_LR))
```

```python
print('Train R^2: ', model_LR.score(X_train, y_train) )
print('Test R^2: ', model_LR.score(X_test, y_test))
```
