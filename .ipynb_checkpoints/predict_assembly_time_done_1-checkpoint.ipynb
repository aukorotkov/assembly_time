### unique_goods_amount (кол-во уникальных товаров в заказе)

Отлогарифмируем данные и применим правило трех сигм, чтобы удалить сильные выбросы

```python
def plot_hist_box(data):
    plt.figure(figsize=(15, 8))
    plt.subplot(2, 1, 1)
    sns.histplot(data=data, bins=45, kde=True)
    plt.grid()

    plt.subplot(2, 1, 2)
    sns.boxplot(data=data, orient='horizontal')
    plt.grid()

    plt.show()
```

```python
#Будем удалять выбросы по правилу трех сигм 
def remove_outliers_iqr(data, multiplier=3):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - multiplier * iqr
    upper_bound = q3 + multiplier * iqr
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    cleaned_data = data[(data >= lower_bound) & (data <= upper_bound)]
    return cleaned_data
```

```python
plot_hist_box(df['unique_goods_amount'])
```

```python
unique_goods_amount = np.log1p(df['unique_goods_amount'])
unique_goods_amount = remove_outliers_iqr(unique_goods_amount)

print(f"Процент удаленных данных: {100 - 100 *len(unique_goods_amount) / len(df)}")
```

```python
plot_hist_box(unique_goods_amount)
```

### Модель

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score

from sklearn.pipeline import Pipeline
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import FunctionTransformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
```

```python
numeric = ['unique_goods_amount', 'pieces_goods_amount', 'kg_goods_amount', 'pieces_goods_total', 'kg_goods_weight']
category = ['store_format']
```

```python
feature = df.drop(columns=['assembly_time'], axis=1)
target = df['assembly_time']
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.25, random_state = 42)
```

```python
column_transformer = ColumnTransformer([
    ('category', OneHotEncoder(drop='first'), category), 
    ('numeric', StandardScaler(), numeric)
], remainder='passthrough' )
```

```python
# Определим функцию преобразования log1p для таргета
def log1p_transform(y):
    return np.log1p(y)

# Определим функцию обратного преобразования expm1 для предсказаний
def expm1_transform(y):
    return np.expm1(y)

# Создадим объект FunctionTransformer для преобразования log1p
log_transformer = FunctionTransformer(log1p_transform, inverse_func=expm1_transform)
```

```python
# Определим модель, которую будем использовать
model_linear = LinearRegression()

# Создадим Pipeline с преобразованием таргета
pipeline = Pipeline([
    ('transform', column_transformer),
    ('linear', model_linear)
])

# Обернем Pipeline в TransformedTargetRegressor для автоматического преобразования таргета и предсказаний
regressor = TransformedTargetRegressor(regressor=pipeline, transformer=log_transformer)

# Используем regressor для обучения модели и предсказания
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
linear_test = pd.DataFrame(y_test).reset_index(drop=True).copy()
linear_test['pred'] = y_pred
```

```python
linear_test
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
sns.histplot(data=linear_test, kde=True)
plt.grid()

plt.subplot(2, 1, 2)
sns.boxplot(data=linear_test, orient='horizontal')
plt.grid()

plt.show()
```

```python
print('Test MAE: ', mean_absolute_error(y_test, y_pred))
# print('Test MSE: ', mean_squared_error(y_test, y_pred))
print('Test RMSE: ', mean_squared_error(y_test, y_pred, squared=False))
```

```python
def scatter_predict(data, target, prediction):
    feature_names = data.columns

    #количества строк и столбцов 
    num_rows = 3
    num_cols = (len(feature_names) // num_rows) + 1
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 15))

    axes = axes.flatten()

    for idx, feature in enumerate(feature_names):

        axes[idx].scatter(data[feature], target, color='#48CAE4', label='Обучающая выборка')
        axes[idx].scatter(data[feature], prediction, color='#FCA311', label='Предсказания выборка')
        axes[idx].set_xlabel('Признак')
        axes[idx].set_ylabel('Целевая переменная')
        axes[idx].set_title(f'График рассеяния для {feature}')
        axes[idx].legend()

    if len(feature_names) < num_rows * num_cols:
        for i in range(len(feature_names), num_rows * num_cols):
            fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()
```

```python
scatter_predict(X_test.drop(columns=category, axis=1), y_test, y_pred)
```

### RandomForestRegressor

```python
random_forest = RandomForestRegressor(random_state=42)

params_forest = {'regressor__model__n_estimators':range(110, 130, 5),
                 'regressor__model__min_samples_split':range(2, 3),
                 'regressor__model__min_samples_leaf':range(3, 7),
                 'regressor__model__max_depth':range(4, 16, 2) }
```

```python
column_transformer_tree = ColumnTransformer([
    ('category', OrdinalEncoder(), category), 
    ('numeric', StandardScaler(), numeric)
], remainder='passthrough' )
```

```python
pipe_forest = Pipeline([
    ('transform', column_transformer_tree),
    ('model', random_forest)
])

regressor_forest = TransformedTargetRegressor(regressor=pipe_forest, transformer=log_transformer)
```

```python
search_forest = GridSearchCV(regressor_forest, params_forest, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=2)

search_forest.fit(X_train, y_train);
```

```python
print(f'Best params: {abs(search_forest.best_score_)}')
print(f'Best params: {search_forest.best_params_}')
```

```python
y_pred = search_forest.predict(X_test)
linear_test = pd.DataFrame(y_test).reset_index(drop=True).copy()
linear_test['pred'] = y_pred
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
sns.histplot(data=linear_test, kde=True)
plt.grid()

plt.subplot(2, 1, 2)
sns.boxplot(data=linear_test, orient='horizontal')
plt.grid()

plt.show()
```

```python
print('Test MAE: ', mean_absolute_error(y_test, y_pred))
# print('Test MSE: ', mean_squared_error(y_test, y_pred))
print('Test RMSE: ', mean_squared_error(y_test, y_pred, squared=False))
```

### XGBoost 

```python
model_xgb = xgb.XGBRegressor(random_state=42)

params_xgb = {'regressor__xgb__n_estimators':range(150, 200, 10),
              'regressor__xgb__learning_rate':np.linspace(0.1, 0.4, 4),
              'regressor__xgb__max_depth':range(3, 7) }
```

```python
pipe_xgb = Pipeline([
    ('transform', column_transformer_tree),
    ('xgb', model_xgb) ])

regressor_xgb = TransformedTargetRegressor(regressor=pipe_xgb, transformer=log_transformer)
```

```python
search_xgb = GridSearchCV(regressor_xgb, params_xgb, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=2)

search_xgb.fit(X_train, y_train);
```

```python
print(f'Best score: {abs(search_xgb.best_score_)}')
print(f'Best params: {search_xgb.best_params_}')
```
